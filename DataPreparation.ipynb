```python
#Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
customer_shopping_data_csv = Path("./Resources/source/customer_shopping_data.csv")

customer_shopping_data_df = pd.read_csv(customer_shopping_data_csv)

#conversion to date type
customer_shopping_data_df["invoice_date"] = pd.to_datetime(customer_shopping_data_df["invoice_date"], format="%d/%m/%Y").dt.date

customer_shopping_data_df.head()
```

```python
exchange_rate_csv = Path("./Resources/output/exchange_rate.csv")

exchange_rate_df = pd.read_csv(exchange_rate_csv)

#conversion to date type
exchange_rate_df["date"] = pd.to_datetime(exchange_rate_df["date"]).dt.date

exchange_rate_df.head()
```

```python
historical_weather_csv = Path("./Resources/output/Istanbul_historical_weather.csv")

historical_weather_df = pd.read_csv(historical_weather_csv)

#conversion to date type
historical_weather_df["date"] = pd.to_datetime(historical_weather_df["date"]).dt.date

historical_weather_df.head()
```

```python
customer_shopping_data_df = pd.merge(customer_shopping_data_df, exchange_rate_df, how="left", left_on="invoice_date", right_on="date")
customer_shopping_data_df.drop(columns=["date"], inplace=True)

customer_shopping_data_df.head()
```

```python
customer_shopping_data_df = pd.merge(customer_shopping_data_df, historical_weather_df, how="left", left_on="invoice_date", right_on="date")

customer_shopping_data_df.drop(columns=["date"], inplace=True)

customer_shopping_data_df.head()
```

```python
customer_shopping_data_df = customer_shopping_data_df.rename(columns={
    "invoice_no": "Invoice #",
    "customer_id": "Customer ID",
    "gender": "Gender",
    "age": "Age",
    "category": "Category",
    "quantity": "Quantity",
    "price": "Price (TYR)",
    "payment_method": "Payment Method",
    "invoice_date": "Invoice Date",
    "shopping_mall": "Shopping Mall",
    "exchange_rate": "Exchange Rate (USD-to-TYR)",
    "tempmax": "Max Temp (F)",
    "tempmin": "Min Temp (F)",
    "temp": "Temp (F)",
    "feelslikemax": "Feels Like Max Temp (F)",
    "feelslikemin": "Feels Like Min Temp (F)",
    "feelslike": "Feels Like Temp (F)",
    "dew": "Dew Point Temp (F)",
    "humidity": "Humidity (%)",
    "precip": "Precip (in)",
    "precipprob": "Precip Likelihood (%)", 
    "precipcover": "Precip Cover (rate)",
    "preciptype": "Precip Type", 
    "snow": "Snowfall (in)",
    "snowdepth": "Snow Depth (in)", 
    "windgust": "Wind Gusts (mph)", 
    "windspeed": "Wind Speed (mph)", 
    "winddir": "Wind Direction (degrees)", 
    "pressure": "Pressure (mbar)",
    "cloudcover": "Clouds (%)", 
    "visibility": "Visibility (mi)", 
    "uvindex": "UV Index", 
    "sunrise": "Sunrise", 
    "sunset": "Sunset",
    "conditions": "Conditions", 
    "description": "Description"
})

customer_shopping_data_df.head()
```

```python
#Entering age category using the American system of division into young, adult, and senior
#set the boundaries of the ranges and their labels, in the last one I specified 1000 years, 
#in this case, it represents an equivalent of >= 65

age_group_bins = [0, 18, 65, 1000]
age_group_labels = ["Young", "Adult", "Senior"]

customer_shopping_data_df["Age Group"] = pd.cut(customer_shopping_data_df["Age"], bins=age_group_bins, labels=age_group_labels, right=False)
customer_shopping_data_df.head()
```

```python
#Similarly, we perform the division into age groups.
age_range_bins = [0, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 100]
age_range_labels = ["0-19", "20-24", "25-29", "30-34", "35-39", "40-44", "45-49", "50-54", "55-59", "60-64", "65-100"]

customer_shopping_data_df["Age Range"] = pd.cut(customer_shopping_data_df["Age"], bins=age_range_bins, labels=age_range_labels, right=False)
customer_shopping_data_df.head()
```

```python
#Apply currency conversion to create a new column with price in $, 
#dividing the price in lira by the exchange rate (dividing because we have downloaded the exchange rate of $ to lira)

customer_shopping_data_df["Price (USD)"] = customer_shopping_data_df["Price (TYR)"] / customer_shopping_data_df["Exchange Rate (USD-to-TYR)"]

#Calculate the cost in lira, based on the price and quantity of the purchased product

customer_shopping_data_df["Cost (TYR)"] = customer_shopping_data_df["Price (TYR)"] * customer_shopping_data_df["Quantity"]
customer_shopping_data_df["Cost (USD)"] = customer_shopping_data_df["Price (USD)"] * customer_shopping_data_df["Quantity"]
customer_shopping_data_df.head()
```

```python
column_order = ['Invoice #', 'Invoice Date', 'Customer ID', 'Gender', 'Age', 'Age Group', 'Age Range', 'Category', 'Quantity', 'Price (TYR)', 'Cost (TYR)', 'Price (USD)', 'Cost (USD)',
                'Exchange Rate (USD-to-TYR)', 'Payment Method', 'Shopping Mall', 'Max Temp (F)', 'Min Temp (F)', 'Temp (F)', 'Feels Like Max Temp (F)', 'Feels Like Min Temp (F)',
                'Feels Like Temp (F)', 'Dew Point Temp (F)', 'Humidity (%)', 'Precip (in)', 'Precip Likelihood (%)', 'Precip Cover (rate)', 'Precip Type', 'Snowfall (in)', 
                'Snow Depth (in)', 'Wind Gusts (mph)', 'Wind Speed (mph)', 'Wind Direction (degrees)', 'Pressure (mbar)', 'Clouds (%)', 'Visibility (mi)', 'UV Index', 
                'Sunrise', 'Sunset', 'Conditions', 'Description']

customer_shopping_data_df = customer_shopping_data_df[column_order]
```

```python
#Add date attributes - month number, day number in the year, date in the year (i.e., date in month/day format without the year), year

customer_shopping_data_df["Month"] = customer_shopping_data_df["Invoice Date"].apply(lambda x: x.month)
customer_shopping_data_df["Day Of Year"] = customer_shopping_data_df["Invoice Date"].apply(lambda x: x.timetuple().tm_yday)
customer_shopping_data_df["Date Of Year"] = customer_shopping_data_df["Invoice Date"].apply(lambda x: f"{x.timetuple().tm_mon}/{x.timetuple().tm_mday}")
customer_shopping_data_df["Year"] = customer_shopping_data_df["Invoice Date"].apply(lambda x: f"{x.timetuple().tm_year}")
```

```python
customer_shopping_data_df.to_csv("./Resources/output/customer_shopping_data.csv", header=True, index=False)
```
